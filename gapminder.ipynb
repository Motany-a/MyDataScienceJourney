<h3>Calculating Covariance</h3>

<h5 style="color:red">Unnecessary😂<h5>

```python
col = ['life', 'child_mortality'] 
from sklearn.preprocessing import StandardScaler
stdc = StandardScaler()
Scale_fun = stdc.fit_transform(df[col].iloc[:,range(0,2)].values)
```

```python
datas = np.cov(Scale_fun.T)
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hema = sns.heatmap(datas, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':12}, yticklabels=col, xticklabels=col)
plt.title('Covariance matrix showing coefficients')
plt.tight_layout()
plt.show()
```

<h4>Splitting DataSets</h4>

```python
from sklearn.model_selection import train_test_split
x = df[col].iloc[:,1:2].values
y = df[col]['life']
```

```python
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size = 0.3, random_state = 0 )
```

<h4>Building a Regression Model</h4>

```python
from sklearn.linear_model import LinearRegression
ln = LinearRegression()

ln.fit(x_train, y_train)
```

```python
y_pred = ln.predict(x_test)
```

```python
y_pred
```

```python
ln.score(x_test,y_test)
```

```python
ln.score(x_train, y_train)
```

```python
ln.score(x_test,y_test)
```
